```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size']=34
%matplotlib inline
from collections import Counter
```

```python
battles= pd.read_csv("battles.csv")
character_deaths=pd.read_csv("character-deaths.csv")
character_predictions=pd.read_csv("character-predictions.csv")
```

```python
A=Counter(character_deaths['Book Intro Chapter'])
B=Counter(character_deaths['Book of Death'])
C=Counter(character_deaths['Death Chapter'])
```

```python
font = {'family': 'arial',
        'color':  'Black',
        'weight': 'normal',
        'size': 19,
        }
```

```python
plt.figure(figsize=(9,9))
plt.title("Chapter vs No. of Characters Introduced",fontdict=font)
plt.ylabel('Character',fontdict=font)
plt.xlabel('Chapter',fontdict=font)
plt.bar(A.keys(),A.values())
```

```python
A.keys()
```

```python
plt.figure(figsize=(9,9))
plt.title("Chapter vs No. of Character Deaths",fontdict=font)
plt.ylabel('Character Deaths',fontdict=font)
plt.xlabel('Chapter',fontdict=font)
plt.bar(B.keys(),B.values())
```

```python
plt.bar(C.keys(),C.values())
plt.title("Book vs No. of Character Deaths",fontdict=font)
plt.ylabel('Character Deaths',fontdict=font)
plt.xlabel('Book',fontdict=font)
plt.autoscale()
plt.show()
```

```python
D=Counter(battles['attacker_king'])
E=Counter(battles['defender_king'])
F=Counter(battles['region'])
G=Counter(battles['defender_1'])
```

```python
plt.figure(figsize=(9,9))
plt.pie([float(o) for o in D.values()], labels=[k for k in D.keys()],autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Attacker king',  loc='center',fontdict=font)
```

```python
plt.figure(figsize=(9,9))
plt.pie([float(o) for o in E.values()],labels=[k for k in E.keys()],autopct='%1.1f%%')
plt.axis('auto')
plt.title('Defender King',loc='center',fontdict=font)
```

```python
plt.figure(figsize=(9,9))
plt.pie([float(o) for o in F.values()],labels=[k for k in F.keys()],autopct='%1.1f%%')
plt.axis('auto')
plt.title('Battle Region',loc='center',fontdict=font)
```

```python
plt.figure(figsize=(9,9))
plt.pie([float(o) for o in G.values()],labels=[k for k in G.keys()],autopct='%1.1f%%')
plt.axis('auto')
plt.title('Defender for Battles',loc='center',fontdict=font)
```

```python
H=Counter(character_predictions['alive'])
I=Counter(character_predictions["popularity"])
```

```python
plt.figure(figsize=(9,9))
plt.boxplot(H.values(),0,'bP')
plt.title("Alive",fontdict=font)
plt.show()
```
