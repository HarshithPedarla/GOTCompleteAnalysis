```python
from textblob import TextBlob
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from wordcloud import WordCloud,STOPWORDS
%matplotlib inline
```

```python
text=open('GOT_Reviews.txt').read()
```

```python
text_list=text.split('\n')
```

```python
count_positive=0
count_negative=0
count_neutral=0
for  i in text_list:
    analysis=TextBlob(i)
    if analysis.sentiment.polarity>0:
        print(i)
        print("The above is Positive")
        count_positive+=1
    elif analysis.sentiment.polarity ==0:
        print(i)
        print("The above is Neutral")
        count_neutral+=1
    else:
        print(i)
        print("The above is Negative")
        count_negative+=1
print(count_positive) 
print(count_negative) 
print(count_neutral) 
```

```python
sentiment_list=[count_positive,count_negative,count_neutral]
sentiment_list
mylabel=["Positive","Negative","Neutral"]
```

```python
font = {'family': 'arial',
        'color':  'Black',
        'weight': 'normal',
        'size': 19,
        }
plt.figure(figsize=(9,9))
plt.pie(sentiment_list,labels=mylabel,autopct='%2.2f%%',radius=1)
plt.title("GOT Review Analysis",fontdict=font,loc='center')
plt.show()
```

```python
maskArray=np.array(Image.open("GOT_White.jpg"))
wc=WordCloud(max_words=200,mask=maskArray,background_color='WHITE')
wc.generate(text)
plt.figure(figsize=(30,20),clear=True,edgecolor="black")
plt.axis("off")
plt.imshow(wc)
```
