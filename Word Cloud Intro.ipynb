```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from wordcloud import WordCloud,STOPWORDS
%matplotlib inline
```

```python
import os
os.getcwd()
```

```python
text=open('GOT_Reviews.txt').read()
text
```

```python
wc=WordCloud()
wc.generate(text)
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(wc)
```

```python
wc=WordCloud(stopwords=STOPWORDS)
wc.generate(text)
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(wc)
```

```python
sw=set(STOPWORDS)
sw.add("great")
sw.add("show")
sw
```

```python
wc=WordCloud(stopwords=sw,background_color='pink')   #sw contains "show" & "great" so word cloud doesnt contain it
wc.generate(text)
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(wc)
```

```python
mask=np.array(Image.open("GOT_White.jpg"))
```

```python
wc=WordCloud(stopwords=sw,mask=mask,background_color='WHITE')   #sw contains "show" & "great" so word cloud doesnt contain it
wc.generate(text)
plt.figure(figsize=(30,20))
plt.axis("off")
plt.imshow(wc)
```
